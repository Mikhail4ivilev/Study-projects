```python
import seaborn as sns
import pandas as pd
from datetime import datetime, timedelta

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
%matplotlib inline

palette = plt.get_cmap('Set2')
```

```python
# Загружаем датасеты и корректируем названия столбцов для удобства

courses = pd.read_csv('courses.csv', index_col=0)
courses.columns = ['course_' + title for title in courses.columns]

students = pd.read_csv('students.csv')
students.drop('id_', axis=1, inplace=True)
students.columns = ['student_' + title for title in students.columns]

course_contents = pd.read_csv('course_contents.csv')

progresses = pd.read_csv('progresses.csv')
progresses.rename(columns={'id': 'progress_id'}, inplace=True)

progress_phases = pd.read_csv('progress_phases.csv', parse_dates= ['start_date', 'finish_date'])
```

```python
# Объединяем датасеты в один

data = progress_phases.merge(progresses, on='progress_id')\
                      .merge(course_contents, on=['course_id', 'module_number', 'lesson_number' ], how='left')\
                      .merge(courses, on='course_id', how='left').sort_values(['module_number', 'lesson_number'])
                   
# data[data['student_id'] == '768c2987a744c51ce64a5993a2a94eaf']

data.head()
```

```python
# Проверяем наличие пропущенных значений

for df in data, students:
    print(df.isnull().sum().apply(lambda x: round(x * 100 / df.shape[0])))
    print()
```

В колонке 'finish_date' 5% пропущенных значений. Это может быть связано с тем, что на момент создания выборки урок еще не был пройден. Также имеется 70% неуказанных городов и 42% пропущенных дат рождения.

```python
# Группируем основную информацию по курсам

courses_inf = data.groupby(['course_title', 'module_number', 'module_title']).agg({'lesson_title': 'nunique'}).reset_index()\
                  .merge(data.groupby('course_title').agg({'module_title': 'nunique'})\
                             .join(data.groupby(['course_title', 'module_title']).agg({'lesson_title': 'nunique'})\
                             .groupby('course_title').agg({'lesson_title': 'median'}))\
                             .rename(columns={'module_title': 'module_count', 'lesson_title': 'lesson_median'}),
                              on='course_title', how='left')\
                  .rename(columns={'lesson_title': 'lesson_count'})\
                  .set_index(['course_title', 'module_count','lesson_median', 'module_number', 'module_title'])

# Сохраняем в отдельный список названия курсов

titles = courses_inf.reset_index()['course_title'].unique().tolist()

print('Количество курсов:', len(titles))
courses_inf
```

```python
# Вычисляем возраст студентов

def age(x):
    
    """Принимает дату рождения, возвращает количество лет до текущей даты. В случае ошибки возвращает 0"""
    
    try:
        return datetime.today().year - pd.to_datetime(x).year
    except Exception:
        return 0


students_age = students.dropna().assign(student_age=students['student_birthday'].apply(age)).query('student_age > 0')

functions = ['min', 'max', 'mean', 'median']
age_values = lambda ser: {function: ser.apply(function) for function in functions}

for name, value in age_values(students_age['student_age']).items():
    print('{} - {:.1f}'.format(name, value))
```

Значения минимального и максимального возрастов явно не соответствуют действительности. Посмотрим на распределение возрастов, чтобы оценить количество таких значений.

```python
# Распределение возрастов

_, ax = plt.subplots(figsize=(12, 6))
plt.style.use('seaborn-darkgrid')
sns.despine(left=False, bottom=False, right=True)

sns.histplot(data=students_age,
             x='student_age',
             bins=20,
             alpha=0.5,
             ax=ax,
             edgecolor=".3",
             linewidth=0.5,).set(ylabel=None, xlabel=None)

ax.set_title("Распределение возраста студентов", fontsize=16)

max_age = age_values(students_age['student_age'])['max']

ax.set_xticks([max_age * i / 20 for i in range(0, 21)])
ax.tick_params('x', rotation=45, labelsize=14)
ax.tick_params('y', labelsize=14)
ax.set_xlabel('Возраст', fontsize=14)
ax.set_ylabel('# записей', fontsize=14)


plt.show()
```

Видно, что количество слишком малых и слишком больших возрастов достаточно мало, поэтому попробем применить правило трех сигм, чтобы избавиться от таких значений.

```python
# Применяем правило трех сигм к различным значениям возраста студентов

mean_age = age_values(students_age['student_age'])['mean']

students_age_ceil = mean_age + 3 * students_age['student_age'].std()
students_age_floor = mean_age - 3 * students_age['student_age'].std()

students_age = students_age[(students_age.student_age >= students_age_floor) &
                            (students_age.student_age <= students_age_ceil)]

for name, value in age_values(students_age['student_age']).items():
    print('{} - {:.1f}'.format(name, value))
```

Хоть минимальный возраст в восемь лет не кажется особо правдоподобным, но в целом результат намного лучше.

```python
# Группируем информацию по количеству и возрасту студентов для каждого курса
# Пропущенные значения заполняем медианным возрастом

mean_age = age_values(students_age['student_age'])['median']

students_info = data.dropna().merge(students_age[['student_id', 'student_age']], on='student_id', how='left')\
                             .fillna(mean_age)\
                             .groupby('course_title', as_index=False).agg({'student_id': 'nunique', 'student_age': functions})\
                             .rename(columns={'student_id': 'student_count'})

students_info.columns = [col[0] + col[1] for col in students_info.columns.values]

students_info
```

Получили, что медианный возраст студентов составляет 30-31 год.

```python
# Расширяем данные о курсах, добавляя информацию об их областях

students_info_ext = students_info.sort_values('student_countnunique', ascending=False)\
                             .merge(data[['course_title', 'course_field']], on='course_title')

median_count = students_info['student_countnunique'].median()
```

```python
_, ax = plt.subplots(figsize=(12, 8))
plt.style.use('seaborn-ticks')
sns.despine(left=False, bottom=False, right=True)

sns.barplot(data=students_info_ext,
            y='course_title',
            x='student_countnunique',
            hue='course_field',
            alpha=0.1,
            dodge=False,
            ax=ax).set(ylabel=None, xlabel=None)

ax.axvline(x=median_count, color='navy', label='Median', linewidth=0.8, linestyle='-')
ax.set_xticks([cnt for cnt in range(0, 2001) if cnt % 500 == 0 or cnt == median_count])

ax.set_title("Количество студентов на каждом курсе", fontsize=16)
ax.legend(loc=1, bbox_to_anchor=(1.25, 1), fontsize=14)
ax.tick_params(labelsize=14)

plt.show()
```

```python
# Фильтруем значения в исходном датасете по наличию домашней работы

data_trimmed = data[(data['is_homework'].eq(True))]

# Создаем столбец с месячным интервалом для рассматриваемой даты

data_trimmed['year_month'] = data_trimmed['start_date'].dt.to_period('M')
```

```python
# Создаем датасет с информацией по первой домашней работе для каждого курса

course_hw_first = data_trimmed.groupby('course_title')[['module_number', 'module_title', 'lesson_number']]\
                        .agg('first').reset_index()

merge_columns = ['course_title', 'module_number', 'module_title', 'lesson_number']
```

```python
# Вычислям количество новых студентов для каждого интервала

increase = data_trimmed.merge(course_hw_first, on=merge_columns, how='right')\
                       .groupby(['course_title', 'year_month']).agg({'student_id': 'nunique'})\
                       .rename(columns={'student_id': 'new_student_count'})

increase
```

Видно, что данные не покрывают все рассматриваемы периоды. Попробуем дополнить информацию.

```python
# Создаем датасет с рассматриваемыми периодами

time_range = pd.date_range('2016-03-01', '2019-07-31', freq='M').to_period('M')
periods = pd.DataFrame({'year_month': [date  for _ in range(len(titles)) for date in time_range]})

periods['course_title'] = [title for title in titles for _ in range(time_range.size)]

# Объединяем  датасеты, пропуски заполняем нулями

new_students = increase.reset_index().merge(periods, on=['course_title', 'year_month'], how='right')\
                        [['course_title', 'year_month', 'new_student_count']].fillna(0)

new_students
```

```python
# Преобразуем значения из столбца с периодами

def month_periods(x):
    
    """Принимает строку вида 'год-номер месяца', возвращает строку вида 'год-название месяца'"""
    
    x = x.split('-')
    
    months_txt = ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec')
    months_ticks = {i: months_txt[i - 1] for i in range(1, 13)}
    
    x[1] = months_ticks[int(x[1])]
        
    return '-'.join(x)

new_students['year_month'] = new_students['year_month'].astype(str).apply(month_periods)  
```

```python
def time_period(vec):
    
    """Принимает на вход начальную конечную даты, возвращает временной ряд с шагом в один месяц. Если разница между
       начальной и конечными датами меньше месяца возвращает дату в  формате 'год-месяц'
    """
    
    if vec[0].to_period('M') == vec[1].to_period('M'):
        return vec[0].to_period('M')
    
    return pd.date_range(vec[0].date(), vec[1].date(), freq='M').to_period('M')

# Незаконченные работы будем считать как нагрузку за все периоды

progresses = data_trimmed[['course_title', 'start_date', 'finish_date']].fillna(pd.to_datetime('2019-07-31'))

progresses['year_month'] = progresses[['start_date', 'finish_date']].apply(time_period, axis=1)

progresses = progresses.explode('year_month').groupby(['course_title', 'year_month']).count()\
                        ['start_date'].reset_index().rename(columns={'start_date': 'hw_count'}).sort_values('year_month')

# Дополняем периодами без домашних работ и заполняем соответсвующие значения нулями

total_progresses = progresses.merge(periods, on=['course_title', 'year_month'], how='right')\
                        [['course_title', 'year_month', 'hw_count']].fillna(0)

# Преобразуем значения из столбца с периодами

total_progresses['year_month'] = total_progresses['year_month'].astype(str).apply(month_periods)  

total_progresses
```

Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

В качестве метрики для проблемных студентов примем разницу во времени между выполнением текущего модуля и началом выполнения следеющего. Студенты с низкой мотивацией обычно не склонны заниматься регулярно. Также можно учесть количество записей 'fail' в каждом модуле. Можно предположить, что студенты с недостаточной начальной подготовкой будут ошибаться чаще других.

```python
# Количество записей 'Fail' для каждого студента

fail = data.groupby(['course_title', 'student_id', 'status']).agg({'status': 'count'}).unstack('status').fillna(0)\
            [('status', 'fail')].reset_index()

# Время между окончанием предыдущего модуля и началом следущего

time_break = data_trimmed.merge(module_hw, on=merge_columns, how='right')\
                       .dropna().groupby(['course_title','student_id', 'module_number', 'module_title'])\
                       .agg({'start_date': 'first', 'finish_date': 'last'})


time_break[['prev_start_date', 'prev_finish_date']] = time_break.groupby(level=1)[['start_date', 'finish_date']].shift(1)
time_break['diff'] = (time_break['start_date'] - time_break['prev_finish_date']).dt.days

# Медианное время для каждого студента, исключая отрицательные и пропущенные записи в колонке 'diff'

time_break = time_break[time_break['diff'] >= 0].groupby(['course_title', 'student_id'])\
                                                .agg({'diff': 'median'}).reset_index()


# Объединяем данные

total_students = time_break.merge(fail, on='student_id', how='inner')\
                        .set_index(['course_title', 'student_id'])\
                        .groupby(level=0, group_keys=False)\
                        .apply(lambda x: x.sort_values(['diff', ('status',  'fail')], ascending=False))\
                        .drop(('course_title',''), axis=1)

# Выбираем студентов со значением 'diff' большим чем среднее на одно стандартное отклонение

problem_students = total_students.groupby(level=0, group_keys=False)\
                        .apply(lambda x: x[x['diff'] > x['diff'].mean() + x['diff'].std()])

problem_students
```
